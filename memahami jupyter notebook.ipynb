# PYTHON DICTIONARY

```python
MLB_team = {
    'Colorado': 'Rockies',
    'Boston': 'Red Sox',
    'Minnesote': 'Twins',
    'Milwaukee': 'Brewers',
    'Seattle': 'Mariners',
}
```

```python
print(MLB_team['Minnesote'])
print(MLB_team['Colorado'])
```

```python
MLB_team['kansas City'] = 'Royals'
MLB_team

{'Colorado': 'Rockies',
 'Boston': 'Red Sox',
 'Minnesote': 'Twins',
 'Milwaukee': 'Brewers',
 'Seattle': 'Mariners',
 'Kansas City': 'Royals'}
```

```python
MLB_team['Seattle'] = 'Seahawks'
MLB_team

{'Colorado': 'Rockies',
 'Boston': 'Red Sox',
 'Minnesote': 'Twins',
 'Milwaukee': 'Brewers',
 'Seattle': 'Seahawks',
 'Kansas City': 'Royals'}
```

# Building a Dictionary Incrementally

```python
person = {}
type(person)

person['fname'] = 'Hack'
person['lname']= 'PTP'
person['age'] = 51
person['spouse'] = 'Edna'
person['children'] = ['Ralph', 'Betty', 'Joey']
person['pets'] = {'dog', 'Fido', 'cat', 'Sox'}
```

```python
person

{'fname': 'Hack',
 'lname': 'PTP',
 'age': 51,
 'spouse': 'Edna',
 'children': ['Ralph', 'Betty', 'Joey'],
 'pets': {'dog': 'Fido', 'cat': 'Sox'}}
```

```python
print(person['fname'])
print(person['lname'])
```

```python
print(person['children'])
print(person['children'][1])
```
